## Step1: Retrieving Responses

* Get subscription key from microsoft account
* Paste subsciption key in headers
* Define params
* Request a url

```
import json
import requests

headers = {
    'Ocp-Apim-Subscription-Key': '308232e62d1d4fc5b0930ee5322c354e',
}

params = ({
    
    #q: key word
    'q': 'lori loughlin',
    
    #count: number of articles to retrieve
    'count': '100',
    
    #offset: first number of articles to skip before retrieving list
    'offset': '0',
    
    #mkt: The region to focus while retrieving articles
    'mkt': 'en-us',
    
    #safesearch: A filter used to filter results for adult content.
    'safesearch': 'Strict',
})

try:
  
    #Request a response by providing url, subscription key and cpnstraints
    response = requests.get("https://api.cognitive.microsoft.com//bing/v7.0/news/search?%s?", headers=headers, params=params)
    
    #Check for error
    response.raise_for_status()
    
    print(response)

#Deal with the exception if one is raised
except Exception as e:
    print("[Errno {0}] {1}".format(e.errno, e.strerror))
    
    # can write "pass" after error to mean that program should be continued
```

## Step 5: Create data frame using articles

```
import pandas as pd
from newspaper import Article

details_np = pd.DataFrame(columns=["url", "authors", "publish_date", "text", "keywords", "summary"])

for link in urls:
  details = []

  try: 
#Add the url
    details.append(link)

#Give article link to newspaper in python
    list_articles = Article(link)

#Download the article
    list_articles.download()

#Parse the article
    list_articles.parse()

#Add name of the authors
    details.append(list_articles.authors)

#Add publish date
    details.append(str(list_articles.publish_date))

#Add article text
    details.append(list_articles.text)

#Implementing Natural Language Processing
    list_articles.nlp()

#Idetifying keywords from article
    details.append(list_articles.keywords)

#Summarize article
    details.append(list_articles.summary)

#Search for a key using a tab after .
#list_articles.

    details_np.loc[len(details_np)] = details
  
  # Include error statement in case - I want it to pass and continue
  except Exception as e:
    pass

details_np
```

## Step 6: Export to csv file
